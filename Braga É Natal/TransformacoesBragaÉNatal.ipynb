```python
import pandas as pd
from datetime import datetime


df_evento = pd.read_csv("csv/Evento.csv")
df_local = pd.read_csv("csv/Local.csv")
df_participante = pd.read_csv("csv/Participante.csv")
df_festival = pd.read_csv("csv/Festival.csv")

df_evento = df_evento.merge(
    df_participante,
    on='participacao',
    how='left'  # para manter todos os eventos
)

df_evento.drop(columns=['participacao'], inplace=True)

df_evento = df_evento.merge(
    df_local[['id_local', 'palco', 'local']],
    on=['palco', 'local'],
    how='left'
)

df_evento.drop(columns=['palco', 'local'], inplace=True)

df_evento = df_evento.drop_duplicates()

df_evento['id_festival'] = 1
df_evento['id_evento'] = range(1, len(df_evento) + 1)

# Passar para YYYY-MM-DD
def normalizar_data(data):
    try:
        # remover espaços e converter
        return datetime.strptime(data.strip(), "%m/%d/%Y").strftime("%Y-%m-%d")
    except Exception:
        return data

# Aplicar aos atributos
df_evento["data_inicio"] = df_evento["data_inicio"].apply(normalizar_data)
df_evento["data_fim"] = df_evento["data_fim"].apply(normalizar_data)
df_festival["data_inicio"] = df_festival["data_inicio"].apply(normalizar_data)
df_festival["data_fim"] = df_festival["data_fim"].apply(normalizar_data)

def obter_ids_eventos_local(id_local):
    ids = df_evento[
        df_evento['id_local'].astype(str).str.contains(rf'\b{id_local}\b', regex=True)
    ]['id_evento']
    return ', '.join(ids.astype(str)) if not ids.empty else ''

df_local['id_evento'] = df_local['id_local'].apply(obter_ids_eventos_local)

def obter_ids_eventos_participante(id_participante):
    ids = df_evento[
        df_evento['id_participante'].astype(str).str.contains(rf'\b{id_participante}\b', regex=True)
    ]['id_evento']
    return ', '.join(ids.astype(str)) if not ids.empty else ''

df_participante['id_evento'] = df_participante['id_participante'].apply(obter_ids_eventos_participante)

df_evento.to_csv("csv/Evento.csv", index=False)
df_local.to_csv("csv/Local.csv", index=False)
df_participante.to_csv("csv/Participante.csv", index=False)
df_festival.to_csv("csv/Festival.csv", index=False)
```
